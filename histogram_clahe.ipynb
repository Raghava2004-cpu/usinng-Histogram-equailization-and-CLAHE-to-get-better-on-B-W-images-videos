```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

image = cv.imread("C:/Users/ragha/Downloads/hill.jpg" , cv.IMREAD_GRAYSCALE)
image = cv.resize(image , (1100,650))

he = cv.equalizeHist(image)
cv.imshow("Histogram equilization" , he)

clahe = cv.createCLAHE(clipLimit=3.0 , tileGridSize= (8,8))
clahe_image = clahe.apply(image)
cv.imshow("CLAHE Image" , clahe_image)

cv.imshow("Original IMage" , image)
cv.waitKey(0)
cv.destroyAllWindows()

plt.figure(figsize= (30,15))
plt.subplot(3,1,1)
plt.imshow(image , cmap = 'grey')
plt.axis("OFF")
plt.title("ORIGINAL IMAGE")

plt.subplot(3,1,2)
he = cv.cvtColor(he , cv.COLOR_BGR2RGB)
plt.imshow(he)
plt.axis("OFF")
plt.title("AFTER APPLYING THE HISTOGRAM EQUAILIZATION")

plt.subplot(3,1,3)
clahe_image = cv.cvtColor(clahe_image , cv.COLOR_BGR2RGB)
plt.imshow(clahe_image)
plt.axis("OFF")
plt.title("AFTER APPLYING THE CLAHE ")
```

```python
import cv2 as cv

video = cv.VideoCapture("C:/Users/ragha/Downloads/black and white video.mp4")

while True:
    ret, frame = video.read()

    if not ret:
        break
    frame = cv.resize(frame, (2000, 740))
    gray_frame = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)
    clahe = cv.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
    frame = clahe.apply(gray_frame)
    frame = cv.resize(frame , (1000,500))
    cv.imshow("Video", frame)
  
    plt.imshow(frame , cmap = 'grey')
    plt.axis("OFF")

    if cv.waitKey(30) & 0xFF == ord("q"):
        break

video.release()
cv.destroyAllWindows()
```
